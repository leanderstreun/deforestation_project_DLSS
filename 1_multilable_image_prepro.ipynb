```python
import os
from skimage import io
import numpy as np

label_folder = "black_labels/"

# Loop through each file and process the image
for filename in os.listdir(label_folder):
    filepath = os.path.join(label_folder, filename)
    
    if "Ice" in filename:
        print(filename)
        houses_mask = io.imread(filepath)
        houses_mask = np.where(houses_mask > 0, 1, 0)
        save_path = os.path.join(label_folder, filename)
        io.imsave(save_path, houses_mask.astype(np.uint8) * 250)
     
    elif "Forest" in filename:
        print(filename)
        roads_mask = io.imread(filepath)
        roads_mask = np.where(roads_mask > 0, 1, 0)
        save_path = os.path.join(label_folder, filename)
        io.imsave(save_path, roads_mask.astype(np.uint8) * 251)
        
    elif "Water" in filename:
        print(filename)
        water_mask = io.imread(filepath)
        water_mask = np.where(water_mask > 0, 1, 0)
        save_path = os.path.join(label_folder, filename)
        io.imsave(save_path, water_mask.astype(np.uint8) * 252)
        
    elif "Land" in filename:
        print(filename)
        land_mask = io.imread(filepath)
        land_mask = np.where(land_mask > 0, 1, 0)
        save_path = os.path.join(label_folder, filename)
        io.imsave(save_path, land_mask.astype(np.uint8) * 253)
        
    elif "Human" in filename:
        print(filename)
        vegetation_mask = io.imread(filepath)
        vegetation_mask = np.where(vegetation_mask > 0, 1, 0)
        save_path = os.path.join(label_folder, filename)
        io.imsave(save_path, vegetation_mask.astype(np.uint8) * 254)
```

```python
import os
from skimage import io
import numpy as np
from matplotlib import pyplot as plt

def apply_color(mask, color):
    """
    Apply a given color to the mask.
    """
    colored_mask = np.zeros((*mask.shape, 3), dtype=np.uint8)
    colored_mask[mask == 1] = color
    return colored_mask

# Define the RGB color values
ice_color = np.array([60, 16, 152])  # '#3C1098'
human_color = np.array([110, 193, 228])  # '#6EC1E4'
water_color = np.array([226, 169, 41])  # '#E2A929'
unlabeled_color = np.array([155, 155, 155])  # '#9B9B9B'
land_color = np.array([132, 41, 246])  # '#8429F6'
forest_color = np.array([254, 221, 58])  # '#FEDD3A'

label_folder = "black_labels/"  # Folder path
tasks = {}

# Group files by task number
for filename in os.listdir(label_folder):
    if filename.endswith(".png"):
        task_num = filename.split('-')[1]
        if task_num not in tasks:
            tasks[task_num] = []
        tasks[task_num].append(filename)

# Print the grouped tasks
print("Grouped tasks:", tasks)

# Process each task group
for task_num, files in tasks.items():
    ice_masks = []
    human_masks = []
    water_masks = []
    land_masks = []
    forest_masks = []

    for filename in files:
        filepath = os.path.join(label_folder, filename)
        mask = io.imread(filepath, as_gray=True)
        label_type = filename.split('-')[-2].lower()  # Extract the label type and convert to lower case for consistency

        print(f"Processing file: {filename}, label_type: {label_type}")

        # Convert to binary mask
        binary_mask = np.where(mask > 0, 1, 0)
        
        if label_type == "water":
            water_masks.append(binary_mask)
        elif label_type == "human":
            human_masks.append(binary_mask)
        elif label_type == "ice":
            ice_masks.append(binary_mask)
        elif label_type == "land":
            land_masks.append(binary_mask)
        elif label_type == "forest":
            forest_masks.append(binary_mask)
    
    # Initialize the combined mask with zeros
    base_shape = None
    if water_masks:
        base_shape = water_masks[0].shape
    elif human_masks:
        base_shape = human_masks[0].shape
    elif ice_masks:
        base_shape = ice_masks[0].shape
    elif land_masks:
        base_shape = land_masks[0].shape
    elif forest_masks:
        base_shape = forest_masks[0].shape
    else:
        print(f"No masks found for task {task_num}")
        continue  # Skip this task if no masks are found

    combined_mask = np.zeros((*base_shape, 3), dtype=np.uint8)

    # Apply the masks in order of priority
    for mask, color in [(water_masks, water_color), (human_masks, human_color), (ice_masks, ice_color), (land_masks, land_color), (forest_masks, forest_color)]:
        for m in mask:
            colored_mask = apply_color(m, color)
            combined_mask = np.where((m[..., np.newaxis] == 1) & (combined_mask == 0).all(axis=-1, keepdims=True), colored_mask, combined_mask)

    # Apply the unlabeled color to any remaining black pixels
    unlabeled_pixels = (combined_mask == 0).all(axis=-1)
    combined_mask[unlabeled_pixels] = unlabeled_color

    # Save the combined mask
    output_path = os.path.join(label_folder, f"image_part_00{task_num}.png")
    io.imsave(output_path, combined_mask)

    # Optionally, display the mask
    plt.imshow(combined_mask)
    plt.title(f'Mask for Task {task_num}')
    plt.axis('off')
    plt.show()

    print(f"Saved combined mask for task {task_num} to {output_path}")
```
